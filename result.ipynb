```python
import json
import pickle

with open('dataframes.pkl', "rb") as pkl_file:
    loaded_dataframes_dict = pickle.load(pkl_file)
```

```python
all_results = []
for idx, (key, values) in enumerate(loaded_dataframes_dict.items()):
    # print(key)
    results = values.copy()
    results['cell'] = key

    all_results.append(results)

print(idx)
```

```python
import pandas as pd

pd.concat(all_results).reset_index().rename(columns={'index':'model_name'}).groupby(by='model_name').mean().sort_values(by='rmse')
```
